```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/BAJFINANCE.csv")
data
```

```python
data = data.drop(columns="Symbol")
data = data.drop(columns="Series")
data = data.drop(columns="Trades")
data = data.drop(columns="Deliverable Volume")
data = data.drop(columns="%Deliverble")
```

```python
data.set_index(pd.DatetimeIndex(data['Date']), inplace=True)
```

```python
data = data[['Close']]
data
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['Close']], data[['EMA_10']], test_size=.2)
```

```python
print(X_test.describe())
```

```python
print(X_train.describe())
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model1 = MLPRegressor(random_state=1, max_iter=500)
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Coefficient of Determination:", r2_score(y_test, y_pred))
```

```python
preds = model1.predict(X_train)
rms=np.sqrt(np.mean(np.power((np.array(y_train)-np.array(preds)),2)))
rms
```
